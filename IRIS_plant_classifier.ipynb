```python
#importing Libraries

from sklearn import datasets
from sklearn import metrics
from sklearn import tree
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load Iris dataset
iris = datasets.load_iris()
x = iris.data
y = iris.target

sc = StandardScaler()
x = sc.fit_transform(x)
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=0)

iris
```

```python
#create a random forest object

model = RandomForestClassifier()

#train model
model.fit(xtrain, ytrain)


print("Train accuracy score: ",metrics.accuracy_score(ytrain,model.predict(xtrain)))

print("Train Confusion metrix: ",metrics.confusion_matrix(ytrain,model.predict(xtrain)))
print("\n")

print("Test Accuracy Score: ",metrics.accuracy_score(ytest,model.predict(xtest)))
cm = metrics.confusion_matrix(ytest,model.predict(xtest))
sns.heatmap(cm, annot=True)
plt.show()
print("\nClassification Report:\n\n ",metrics.classification_report(ytest,model.predict(xtest)))
```
